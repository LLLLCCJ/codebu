```python
class Node:
    def __init__(self, data):
        self.data = data
        self.p = self.left = self.right = None
        self.color = "R"


class RB_tree:
    def __init__(self):
        self.root = None
        self.inserted_node = None

    def _insert(self, data):
        self.root = self.insert_node(self.root, data, None)
        self._balance(self.inserted_node)

    def insert_node(self, node, data, p):
        if node is None:
            node = Node(data)
            node.p = p
            self.inserted_node = node
        else:
            if data < node.data:
                node.left = self.insert_node(node.left, data, node)
            elif data > node.data:
                node.right = self.insert_node(node.right, data, node)
        return node

    def _balance(self, node):
        p = node.p
        if p is None:
            node.color = "B"
        else:
            if p.color == "R":
                g = p.p
                u = None
                if g.left == p:
                    u = g.right
                elif g.right == p:
                    u = g.left
                if u and u.color == "R":
                    u.color, p.color = "B", "B"
                    g.color = "R"
                    
                    self._balance(g)
                else:
                    if g.left == p and p.left == node:
                        g.color, p.color = p.color, g.color
                        self.LL(g)
                    elif g.left == p and p.right == node:
                        self.RR(p)
                        node.color, g.color = g.color, node.color
                        self.LL(g)
                    elif g.right == p and p.right == node:
                        g.color, p.color = p.color, g.color
                        self.RR(g)
                    elif g.right == p and p.left == node:
                        self.LL(p)
                        g.color, node.color = node.color, g.color
                        self.RR(g)

    def LL(self, node):
        c = node.left
        p = node.p

        if c.right:
            c.right.p = node
        node.left = c.right
        node.p = c
        c.right = node
        c.p = p
        if p is None:
            self.root = c
        elif p:
            if p.left == node:
                p.left = c
            elif p.right == node:
                p.right = c

    def RR(self, node):
        c = node.right
        p = node.p
        if c.left:
            c.left.p = node

        node.right = c.left
        node.p = c
        c.left = node

        c.p = p
        if p is None:
            self.root = c
        elif p:
            if p.left == node:
                p.left = c
            elif p.right == node:
                p.right = c
```

```python
def check(node):
    if node.left is not None:
        check(node.left)
    if node.p != None:
        print(
            f'data: {node.data}, parent: {node.p.data}, color: {node.color}')
    else:
        print(f'data: {node.data}, root: {node.p}, color: {node.color}')
    if node.right is not None:
        check(node.right)
```

```python
rbt = RB_tree()
a = [2, 1, 8, 9, 7, 3, 6, 4, 5]
 
for x in a:
    rbt._insert(x)
 
check(rbt.root)
```
