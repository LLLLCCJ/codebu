```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=self.right=self.p=None
        self.color="R"
class RB_tree:
    def __init__(self):
        self.root=None
        self.inserted_node=None
    def _insert(self,data):
        self.root=self._find(self.root,data,None)
        self.balance(self.inserted_node)
    def _find(self,node,data,p):
        if node is None:
            node=Node(data)
            node.p=p
            self.inserted_node=node
        elif node.data>data:
            node.left=self._find(node.left,data,node)
        elif node.data<data:
            node.right=self._find(node.right,data,node)
        return node
    def balance(self,node):
        p=node.p
        if p is None:
            node.color="B"
        else:
            if p.color=="R":
                g=node.p.p
                u=None
                if g.left==p:
                    u=g.right
                elif g.right==p:
                    u=g.left
                if u and u.color==p.color:
                    u.color,p.color="B","B"
                    g.color="R"
                    return self.balance(g)
                else:
                    if p==g.left and p.left==node:
                        g.color,p.color=p.color,g.color
                        self.LL(g)
                    elif p==g.right and g.right==node:
                        self.RR(g)
                        g.color,p.color=p.color,g.color
                    elif p==g.left and p.right==node:
                        g.color,node.color=node.color,g.color
                        self.RR(p)
                        self.LL(g)
                    elif p==g.right and p.left==node:
                        g.color,node.color=node.color,g.color
                        self.LL(p)
                        self.RR(g)

    def LL(self,node):
        p=node.p
        c=node.left
        node.left=c.right
        if c.right:
            c.right.p=node
        c.right=node
        node.p=c
        c.p=p
        if p:
            if p.left==node:
                p.left=c
            elif p.right==node:
                p.right=c
        else:
            self.root=c
        
    def RR(self,node):
        p=node.p
        c=node.right
        node.right=c.left
        if c.left:
            c.left.p=node
        c.left=node
        node.p=c
        c.p=p
        if p:
            if p.left==node:
                p.left=c
            elif p.right==node:
                p.right=c
        else:
            self.root=c
    def show(self,node):
        print(node.color)
        if node.left:
            self.show(node.left)
        if node.right:
            self.show(node.right)
    def _del(self,data):
        node=self.find_del_node(self.root,data)
        if node is None:
            return False
        self.del_node(self,node)
        return True
    def find_del_node(self,node,data):
        if node.data>data:
            return self.find_del_node(node.left,data)
        elif node.data<data:
            return self.find_del_node(node.right,data)
        elif node.data==data:
            L=node.left
            R=node.right
            if L and R:
                pointer=self.find_min_pointer(node.right)
                node.data,pointer.data=pointer.data,node.data
                return pointer
            return node
    def del_node(self,node):
        c,s=None,None
        p=node.p
        if node.left:
            c=node.left
        elif node.right:
            c=node.right
        if p is None:
            self.root=None
            return
        
```
