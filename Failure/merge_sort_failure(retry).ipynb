```python
def merge(A, B):
    merge_List=[]
    p = 0
    q = 0
    while not (p == len(A) or q == len(B)):
        if A[p] > B[q]:
            merge_List.append(B[q])
            print("P>Q : p",p,"q",q)
            print("mergeL :",merge_List)
            q += 1
            print("q :",q)
        else:
            merge_List.append(A[p])
            print("P>Q : p", p, "q", q)
            print("mergeL :",merge_List)
            p += 1
            print("p :", p)
    merge_List += A[p:]
    print("mergeL :",merge_List)
    merge_List += B[q:]
    print("mergeL :",merge_List)
    return merge_List

class Merge_C:
    def __init__(self):
        self.A = Tree()
        self.B = Tree()
        self.count = 0
        self.save_List=Tree()
    def merge_sort(self,x,a,b):
        n=len(x)
        global n
        if len(x)!=1:
            c = (a + b) // 2
            self.A._append(x[:c])
            self.B._append(x[c:])
            print("a", a)
            print("b", b)
            print("c", c)
            print( self.A._allshow())
            print( self.B._allshow())
            print("merge a 출력")
            self.merge_sort(self.A.f_return().data,a,c)
            print("merge b 출력")
            self.merge_sort(self.B.f_return().data,c,b)
        self._Lsave(self.A.f_return(),self.B.f_return().data)
    def _Lsave(self,A,B):
        if self.save_List.f_return().level>1:
            if self.count//2==0
                curr=self.save_List.f_return()
                self.save_List._append(merge(curr.data,curr.next.data))
                print("aList", self.save_List)
            else:
                curr = self.save_List.f_return()
                for i in range(2):
                    curr=curr.next
                self.save_List._append(merge(curr.data,curr.next.data))
                print("bList", self.save_List)
        else:
            self.save_List._append(merge(A,B))
        self.count += 1
        print("List", self.save_List)
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        self.level=None
class Tree:
    def __init__(self):
        self.head=Node(None)
        self.cnt=0
        self.recnt=0
    def _append(self,data):
        newNode=Node(data)
        newNode.next = self.head.next
        self.head.next = newNode
        self.cnt+=1
        self.level=len(newNode)
    def _return(self):
        self.recnt += 1
        curr=self.head
        if self.recnt<=self.cnt:
            for i in range(self.recnt):
                curr=curr.next
            return(curr.data)
    def f_return(self):
        return self.head.next
    def _show(self):
        curr=self.head
        for i in range(self.recnt):
            curr=curr.next
            print(curr.data,end=' ')
    def _allshow(self):
        curr=self.head
        for i in range(self.cnt):
            curr=curr.next
            print(curr.data,end=' ')

x = [1, 2, 3, 4, 5, 6, 7, 8]

a=0
b=len(x)
k=Merge_C().merge_sort(x,a,b)

```
