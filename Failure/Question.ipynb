```python
class Node:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
class B_S_Tree:
    def __init__(self,data):
        self.root=Node(data)
    def _insert(self,data):
        newNode=Node(data)
        node=self._find(self.root,data)
        print("return된 node.data",node.data)
        if node.data<data:
            node.right=newNode
        else:
            node.left=newNode
            print("저장",node.left.data)
    def _find(self,node,data):
        print("find node.data",node.data)

        if node.data<data:
            if node.right is None:
                return node# _find 함수에서 처음 이외에는 return이 되지 않음
            else:
                self._find(node.right,data)
        else:
            if node.left is None:
                print("find return",node.data)
                return node
            else:
                self._find(node.left,data)

x=B_S_Tree(10)
x._insert(5)
x._insert(3)

```
