```python
class Node:
    def __init__(self,data):
        self.data=data
        self.right=self.left=self.p=None
        self.color="R"
class RB_tree:
    def __init__(self):
        self.root=None
        self.inserted_node=None
    def _find(self,data):
        return self._find_data(self.root,data)
    def _find_data(self,root,data):
        if root is None or root.data==data:
            return root
        elif root.data>data:
            return self._find_data(root.left,data)
        elif root.data<data:
            return self._find_data(root.right,data)
    def _insert(self,data):
        self.root=self._insert_node(self.root,data,None)
        self._balancing(self.inserted_node)
        return
    def _insert_node(self,curr,data,p):
        if curr is None:
            curr=Node(data)
            curr.p=p
            self.inserted_node=curr
        else:
            if data<curr.data:
                curr.left=self._insert_node(curr.left,data,curr)
            elif data>curr.data:
                curr.right=self._insert_node(curr.right,data,curr)
        return curr
    def _balancing(self,node):
        p=node.p
        if p is None:
            node.color="B"
        else:
            if p.color=="R":
                g=p.p
                u=None
                if g.left==p:
                    u=g.right
                elif g.right==p:
                    u=g.left
                if u and u.color=="R":
                    p.color,u.color="B","B"
                    g.color="R"
                    self._balancing(g)
                else:
                    if p==g.left and p.left==node:
                        g.color,p.color=p.color,g.color
                        self.LL(g)
                    elif p==g.left and p.right==node:
                        self.RR(p)
                        g.color,node.color=node.color,g.color
                        self.LL(g)
                    elif p==g.right and p.right==node:
                        g.color,p.color=p.color,g.color
                        self.RR(g)
                    elif p==g.right and p.left==node:
                        self.LL(p)
                        g.color,node.color=node.color,g.color
                        self.RR(g)
    def RR(self, node):
        c = node.right
        p = node.p
 
        if c.left is not None:
            c.left.p = node
 
        node.right = c.left
        node.p = c
        c.left = node
        c.p = p
        if p is None:
            self.root = c
        elif p is not None:
            if p.left == node:
                p.left = c
            elif p.right == node:
                p.right = c
 
    def LL(self, node):
        c = node.left
        p = node.p
 
        if c.right is not None:
            c.right.p = node
 
        node.left = c.right
        node.p = c
        c.right = node
        c.p = p
        if p is None:
            self.root = c
        elif p is not None:
            if p.left == node:
                p.left = c
            elif p.right == node:
                p.right = c
```

```python
def check(node):
    if node.left is not None:
        check(node.left)
    if node.p != None:
        print(
            f'key: {node.data}, parent: {node.p.data}, color: {node.color}')
    else:
        print(f'key: {node.data}, root: {node.p}, color: {node.color}')
    if node.right is not None:
        check(node.right)
```

```python
a = [2, 1, 8, 9, 7, 3, 6, 4, 5]
x=RB_tree()
for i in a:
    x._insert(i)
check(x.root)
```
