```python
import pandas as pd
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('seaborn')
sns.set(font_scale=2.5)
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df_train=pd.read_csv('C:/Users/cod/Desktop/ML/train.csv')
df_test=pd.read_csv('C:/Users/cod/Desktop/ML/test.csv')
```

```python
msno.bar(df_train)
```

```python
for col in df_train.columns:
    x=100*(df_train[col].isnull().sum()/df_train[col].shape[0])
    if x>0:
        print('{:<10} : {:.2f}% '.format(
            col,x))
    
```

```python
for col in df_test.columns:
    x=100*(df_test[col].isnull().sum()/df_test[col].shape[0])
    if x>0:
        print('{:<10} : {:.2f}%'.format(col,x))
```

```python
for col in df_train.columns:
    x=100*(df_train[col].isnull().sum()/df_train[col].shape[0])
    if x>40:
        df_train.drop(col,axis=1,inplace=True)
```
