```python
class Node:
    def __init__(self,data):
        self.data=data
        self.right=self.left=self.p=None
        self.color="R"
class RB_tree:
    def __init__(self):
        self.root=None
        self.inserted_node=None
    def _insert(self,data):
        self._find(self.root,data,None)
        self.color_balancing(self.inserted_node)
        return
    def _find(self,node,data,p):
        if node is None:
            node=Node(data)
            node.p=p
            if node.p is None:
                self.root=node
            self.inserted_node=node
        else:
            if node.data>data:
                node.left =self._find(node.left,data,node)
            elif node.data<data:
                node.right= self._find(node.right,data,node)
        return node
    def color_balancing(self,node):
        if node.p is None:
            node.color="B"
        else:
            if node.p.color=="R":
                g=node.p.p
                p=node.p
                u=None
                if g.left==p:
                    u=g.right
                elif g.right==p:
                    u=g.left
                if u and u.color==node.p.color:
                    u.color,node.p.color="B","B"
                    g.color="R"
                    return self.color_balancing(g)
                    #########################
                else:
                    if p==g.left:
                        if p.right==node:
                            p=self.RR(p)
                        g=self.LL(g)
                        g.color,g.right.color=g.right.color,g.color
                    elif p==g.right:
                        if p.left==node:
                            self.LL(p)
                        g=self.RR(g)
                        g.color,g.left.color=g.left.color,g.color

    def LL(self,node):
        p=node.p
        c=node.left
        node.left=c.right
        if c.right:
            c.right.p=node
        c.right=node
        node.p=c
        c.p=p
        if p:            
            if p.left==node:
                p.left=c
            elif p.right==node:
                p.right=c
        else:
            self.root=c
        return c
    def RR(self,node):
        p=node.p
        c=node.right
        node.right=c.left
        c.left=node
        if c.left:
            c.left.p=node
        c.p=p
        if p:
            
            if p.left==node:
                p.left=c
            elif p.right==node:
                p.right=c
        else:
            self.root=c
        return c
    def show(self,node):
        print(node.color,node.data)
        if node.left:
            self.show(node.left)
        if node.right:
            self.show(node.right)
                
```

```python
a = [2, 1, 8, 9, 7, 3, 6, 4, 5]
x=RB_tree()
for i in a:
    x._insert(i)
x.show(x.root)
```
